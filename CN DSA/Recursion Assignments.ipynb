```python
def palindrome(s):
    l = len(s)
    if l == 0 or l == 1:
        return True
    return (s[0] == s[l-1]) and palindrome(s[1:l-1])
```

```python
palindrome("sr")
```

Suppose there are n persons and they are having n chocolates each. We have to calculate total number of chocolates.
What can be the base case i.e when there are no persons so there will be no chocolaes so return 0.
Induction Hypo: We can just calculate the number of chocolates one person having and say recursion to calculate for others(n-1) persons.

```python
def multipication(m,n):
    # base case
    if n == 0:
        return 0
    return m + multipication(m,n-1)
```

```python
def str_int(s):
    l = len(s)
    # base case
    if l == 1: # we can consider if last one is zero than it can work
        return s
    if s[0] == '0':
        return str_int(s[1:])
    else:
        return s
```

```python
str_int('0001230045000')
```

```python
str_int('1234')
```

```python
def stringtoint(s):
    l = len(s)
    if l == 1:
        return ord(s) - ord('0')
    last = ord(s[-1]) - ord('0')
    return (stringtoint(s[:-1]) * 10) + last
```

```python
def pair_star(s):
    l = len(s)
    if l == 1 or l == 0:
        return s
    if s[0] == s[1]:
        return s[0] + '*' + pair_star(s[1:])
    else:
        return s[0] + pair_star(s[1:])
```

```python
pair_star("aabbbccccddddd")
```

# Qus 8
### Check AB

Send Feedback

Suppose you have a string, S, made up of only 'a's and 'b's. Write a recursive function that checks if the string was generated using the following rules:

a. The string begins with an 'a'

b. Each 'a' is followed by nothing or an 'a' or "bb"

c. Each "bb" is followed by nothing or an 'a'

If all the rules are followed by the given string, return true otherwise return false.

Input format :
String S

Output format :
'true' or 'false'

Constraints :
0 <= |S| <= 1000

where |S| represents length of string S.

Sample Input 1 :
abb

Sample Output 1 :
true

Sample Input 2 :
abababa

Sample Output 2 :
false

```python
def check_ab(s):
    l = len(s)
    if l == 1 or l == 0:
        return True
    if (s[0] == 'a') and (s[:2] == 'aa' or s[1:3] == 'bb' ):
        return True and check_ab(s[3:])
    else:
        return False
```

```python
check_ab('abbabb')
```

# Qus 9
### Staircase

Send Feedback

A child is running up a staircase with N steps, and can hop either 1 step, 2 steps or 3 steps at a time. Implement a method to count how many possible ways the child can run up to the stairs. You need to return number of possible ways W.

Input format :
Integer N

Output Format :
Integer W

Constraints :
1 <= N <= 30

Sample Input 1 :
4

Sample Output 1 :
7

Sample Input 2 :
5

Sample Output 2 :
13

```python
def stairs(n):
    if n == 1:
        return 1
    if n == 2:
        return 2
    if n == 3:
        return 4
    return (stairs(n-1) + stairs(n-2) + stairs(n-3))
```
